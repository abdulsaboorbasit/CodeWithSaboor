1. **The 1 to 100:** Write a loop that prints every number from 1 to 100, but only if it's a multiple of 5.

```python
for number in range(0,101,5):
    print(number)
```

2. **The "Are We There Yet?":** Use a `while` loop that keeps asking the user "Can I stop?" and only ends when the user types "Yes."

```python
while True:
    message = input("Can I stop?")
    if message.capitalize() == "Yes":
        break
```

3. **The FizzBuzz Junior:** Loop from 1 to 20. If a number is divisible by 3, print "Fizz." Otherwise, print the number.

```python
for number in range(1,21):
    if number%3 == 0:
        print("Fizz")
    else:
        print(number)
```

4. **Blast Off!:** Use a loop to count down from 10 to 1, then print "🚀 Lift off!"

```python
for number in range(10,0,-1):
    print(number)

print("🚀 Lift off!")
```

5. **The List Doubler:** Take a list of numbers and use a loop to create a *new* list where every number is doubled.

```python
numbers = [2,3,4,5,6]
new_numbers = []
for number in numbers:
    new_numbers.append(number**2)
print(new_numbers)
```
