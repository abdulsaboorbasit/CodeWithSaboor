3. **The Bank Account:** Create a `BankAccount` class with a `balance` attribute. Add a method `deposit(amount)` that adds money to the balance, and another to display the current balance.

```python
class bank_account:
    def __init__(self,balance):
        self.balance = balance
    

    def deposit(self,amount):
        self.balance = self.balance + amount
        print(f"Rs. {amount} is deposited successfully.")

    def current_balance(self):
        print(f"The current balance is {self.balance}")
        
balance = 1000
account = bank_account(balance)

print("<===Menu===>")
print("1. Deposit Money")
print("2. Current Balance")
print("Exit")
while True:
    choice = int(input())
    if choice == 1:
        amount = int(input("Enter Amount to deposit: "))
        account.deposit(amount)
    elif choice == 2:
        account.current_balance()
    elif choice == 3:
        print("The Program Exit Successfully.")
        break
    else:
        print("Invalid Input.")
```

4. **The Student ID:** Create a `Student` class where the `__init__` method takes a `name` and a list of `grades`. Add a method that calculates and returns the student's average grade.

```python
class student:
    def __init__(self):
        self.name = None
        self.grades = []

    def student_data(self):
        self.name = input("Enter Your Name: ")
        while True:

            choice = input("Enter Grades / Type Quit to Exit: ")
            if choice.lower() == "quit":
                break
            try:
                choice = int(choice)
                self.grades.append(choice)
            except:
                choice.lower() != "quit"
                print("Wrong Input")
        print("Student data added successfully.")

    def average(self):
        marks = 0
        for grade in self.grades:
            marks = marks + grade    
        avg_grades = marks / len(self.grades)
        return avg_grades
        

    def display(self):
        if len(self.grades) == 0:
            print("\nNo Data Found!")
        else:
            print(self.name)
            print(self.average())

student_1 = student()
while True:
    print("\n<===Menu===>")
    print("1. Enter Student Data")
    print("2. Display Student Average Marks")
    print("3. Exit")
    choice = int(input())
    if choice == 1:
        student_1.student_data()
    elif choice == 2:
        student_1.display()
    elif choice == 3:
        print("The Program Exit Successfully.")
        break
    else:
        print("Invalid Input.")
            

```

5. **The Rectangle Area:** Create a `Rectangle` class that takes `length` and `width`. Add a method called `calculate_area()` that returns the area of the rectangle.

```python
class rectangle:
    def __init__(self):
        self.length = None
        self.width = None

    def calculate_area(self):
        self.length = float(input("Enter Length: "))
        self.width = float(input("Enter Width: "))
        area = self.length * self.width
        return area
    
rec_area = rectangle()
print(f"The area of rectangle is: {rec_area.calculate_area()}.")
```
