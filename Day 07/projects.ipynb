1. **Inheritance (The Family Tree):** Create a parent class called `Animal` with a method `eat()`. Create a child class called `Cat` that inherits from `Animal` and adds its own method `meow()`. 

```python
class animal:
    def eat(self):
        print("Eating....")

class cat(animal):
    def meow(self):
        print("The cat is meowing.")

cat_1 = cat()
# Inheriting feature of parent class:
cat_1.eat()
# Feature of child class:
cat_1.meow()
```

2. **Polymorphism (Many Forms):** Create two classes, `Dog` and `Cat`. Both should have a method named `make_sound()`. The dog should print "Woof" and the cat should print "Meow". Write a function that takes *any* animal object and calls its `make_sound()` method!

```python
class dog:
    def make_sound(self):
        print("Woof! Woof!")

class cat:
    def make_sound(self):
        print("Meow! Meow!")

# This is Polymorphism 
# Same function name but behaves different

dog_1 = dog()
dog_1.make_sound()

cat_1 = cat()
cat_1.make_sound()
```

3. **Encapsulation (Private Vault):** Create a `Person` class. Make the `age` attribute private by putting two underscores before it (like `self.__age`). Create a method to let the person see their age, but don't let anyone change it directly from outside the class!

```python
class person:
    def __init__(self,age):
        self.__age = age

    def display(self):
        print("Age:", self.__age)

person_1 = person(24)
person_1.display()
# This line of code is to update a private variable outside a class and it is not accessible.
# person_1.__age == 25
person_1.display()
```

4. **The Smart Device (Super() Function):** Create a `Phone` class with a method `Phone()`. Create a `SmartPhone` class that inherits from it. Use the `super()` function to inherit the calling ability, but add a brand new feature like `browse_internet()`.

```python
class phone:
    def __init__(self,is_on):
        self.__is_on = is_on

    def call(self, number):
        if self.__is_on:
            print(f"Dialing {number}...")
        else:
            print("Cannot call. Phone is off.")

    def send_text(self):
        print("Sending message....")

    def power_on(self):
        if not self.__is_on:
            self.__is_on = True
            print(f"Phone is booting up... Welcome!")
        else:
            print("Phone is already on.")
    
    def power_off(self):
        self.__is_on = False
        print("Shutting down... Powering off.") 

class smart_phone(phone):
    def __init__(self, is_on):
        super().__init__(is_on)
    
    def browse_internet(self):
        print("Opening browser... Now you can browse the internet.")

phone_1 = smart_phone(False)
phone_1.power_on()
phone_1.call("15")
phone_1.browse_internet()
```

5. **The Library System:** Create a `Book` class and a `Library` class. The `Library` class should have a list of book objects. Add a method to the library to add books and another to list out all the books currently available.

```python
class book:
    def __init__(self, title):
        self.title = title

class library:
    def __init__(self):
        self.books = []

    def add_books(self):
        while True:
            title = input("Enter Book Title / Quit: ")
            if title.lower() == "quit":
                break
            else:
                new_book = book(title)
                self.books.append(new_book)

    def show_books(self):
        for i in range(len(self.books)):
            print(f"{i + 1}. {self.books[i].title}")

lms = library()
lms.add_books()
lms.show_books()
```
