1. **The Tip Calculator:** Ask the user for a bill amount and a tip percentage. Print the total bill.

```python
bill = float(input("What is my bill?"))
tip = int(input("What is should be tip? 10,15 or 20%?"))
tip_amount = bill*tip/100
total_bill = int(bill + tip_amount)
print(f"Waiter: Sir! Your total bill is Rs.{total_bill}")
```

2. **Age in Seconds:** Take a user's age in years and convert it into roughly how many seconds they've been alive. 

```python
age = int(input("What is your age?"))
age_in_sec = age*365*24*60*60
print(f"You have live about {age_in_sec} seconds.")
```

3. **The Temperature Transformer:** Create a script that converts Celsius to Fahrenheit using the formula: $F = (C \times 9/5) + 32$.

```python
c_temp = float(input("Enter Temperature in Celsius: "))
f_temp = (c_temp*9/5)+32
print(f"The {c_temp} celsius temperature in fahrenheit is {f_temp}.")
```

4. **Variable Swapper:** Create two variables, `a` and `b`. Write code to swap their values without manually re-typing the numbers.

```python
a = 10
b = 20
# Value of a and b before swapping!
print(f"a = {a}")
print(f"b = {b}")
a, b = b, a
# Value of a and b after swapping!
print(f"a = {a}")
print(f"b = {b}")
```
