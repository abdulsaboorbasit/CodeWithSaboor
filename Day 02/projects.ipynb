1. **The Bouncer:** Create a script that asks for an age. If 18+, print "Welcome!"; if under 18, print "Go drink some juice."

```python
age = int(input("What is your age?"))
if age > 18:
    print("Welcome!")
else:
    print("Go drink some juice.")
```

1. **Even or Odd:** Ask the user for a number. Use the modulus operator `%` to tell them if it’s even or odd.

```python
number = int(input("Enter a number: "))
if number%2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

4. **The Secret Password:** Create a variable `secret = "Python123"`. Ask the user for input; if they match it, print "Access Granted."

```python
secret = "Python123"
user = input("Enter Your Password: ")
if secret == user:
    print("Access Granted")
else:
    print("Wrong Password")
```

5. **List Max-Min:** Create a list of 5 numbers. Without using built-in functions like `max()` or `min()`, try to figure out which logic you'd use to find the biggest number.


```python
numbers = [51,2,63,24,15]
max = 51
min = 51
for number in numbers:
    if max > number:
        max = max
    else:
        max = number
print(f"Maximum Number = {max}")
for number in numbers:
    if min < number:
        min = min
    else:
        min = number
print(f"Minimum Number = {min}")
```
