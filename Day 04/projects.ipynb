1. **The Digital Phonebook:** Create a Dictionary where names are keys and phone numbers are values. Ask the user for a name and print their number.

```python
name = input("Enter Your Name: ")
number = input("Enter Your Number: ")
phone_book = {"Name":name,
              "Number":number}
print(phone_book)
```

```python
names = []
numbers = []
phone_book = {"Name":names,
              "Number":numbers}
while True:
    name = input("Enter Your Name / Type Quit to Exit: ")
    if name.lower() == "quit":
        break
    number = input("Enter Your Number: ")
    names.append(name)
    numbers.append(number)
# To print whole Dictionary
print(phone_book)
# To print Single Key-Value Pair from Dict
print(phone_book["Name"][0],phone_book["Number"][0])
```

2. **The Unique Party Guestlist:** Take a list of names that has duplicates. Use a **Set** to automatically remove the duplicates and print the unique names.

```python
guest_list = ["Abdul Saboor","Aqeel Sabir","Muhammad Hashir","Muhammad Salman","Aqeel Sabir"]
guests = set(guest_list)
print(guests)
```

4. **Stock Checker:** Create a dictionary of fruits and their prices. Print a sentence saying: "An apple costs $2."

```python
fruits = {
    "Apple":"2",
    "Banana":"5",
    "Mango":"8"
}
list_fruits = list(fruits.keys())
print(f"An {list_fruits[0]} costs ${fruits['Apple']}")
```

5. **Dictionary Loop:** Loop through your "Phonebook" dictionary and print only the names (the keys).

```python
phone_book = {
    "Abdul Saboor":"0334-1234567",
    "Aqeel Sabir":"0300-1234567",
    "Muhammad Hashir":"0347-1234567",
    "Muhammad Salman":"0324-1234567",
}
for key,value in phone_book.items():
    print(f"Name = {key}, Number = {value}")
```
