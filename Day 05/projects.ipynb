2. **The Calculator Function:** Create a function that takes two numbers and an operator (`+`, `-`, `*`, `/`) and returns the result.

```python
number_1 = int(input("Enter First Number: "))
number_2 = int(input("Enter Second Number: "))
operator = input("Enter an operator (+,-,*,/): ")
def calculate(number_1,number_2,operator):
    if operator == "+":
        result = number_1 + number_2
    elif operator == "-":
        result = number_1 - number_2
    elif operator == "*":
        result = number_1 * number_2
    elif operator == "/":
        if number_2 != 0:
            result = number_1 / number_2
        else:
            print("Error: Division by Zero.")
    else:
        print("Invalid Operator.")
    return result

print(calculate(number_1,number_2,operator))
```

3. **The Crash Proof Input:** Use a `try/except` block to ask the user for a number. If they type a word instead, print "Hey! That's not a number!" instead of crashing.

```python
try:
    number = int(input("Enter a Number: "))
    print(f"The number enter by user is: {number}.")
except:
    print(f"Hey! That's not a number.")
```

4. **The Journal Writer:** Write a script that asks the user for their "Thought of the day" and saves it into a file called `diary.txt`.

```python
message = input("What is your thought of the day?")
with open("diary.txt","w") as file:
    file.write(message)
```

5. **The File Reader:** Write a script that opens `diary.txt` and prints the contents back to the user in all capital letters.

```python
with open("diary.txt","r") as file:
    content = file.read()
print(content)
```
